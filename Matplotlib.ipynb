- Most important feature is its ability to run well with many operating systems and graphics backends


- The best use of Matplotlib differs depending on how you are using it .i.e., In a script, in Ipython terminal
  or in an IPython notebook

**%matplotlib** - embed interactive plots within the notebook

**%matplotlob inline** - embed static images of your plot in the notebook (once per kernel/session)
                         (any cell within the notebook that creates a plot will embed a PNG image of the resulting
                         graphic)

```python
%matplotlib inline
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

-  The **figure** (an instance of the class plt.Figure) can be thought of as a single container that containd 
    all the objects representing axes, graphics, text, labels
    
    
- The **axes**(an instance of the class plt.Axes) is a bounding box with ticks and labels, which will eventually
   contain the plot elements that make up our visualization

### Two interfaces to create plots

#### MATLAB-style interface:

```python
x = np.linspace(0,10, 100)
```

```python
plt.figure()  # create a plot figure #stateful object (remembers figure and axis)

# create the first two panels and set current axis
plt.subplot(2,1,1) #(rows, columns, panel number)
plt.plot(x, np.sin(x));

#create the secind panel and set current axis
plt.subplot(2,1,2) 
plt.plot(x, np.cos(x));
```

```python
#Saving Figures to File:

fig = plt.figure()

plt.plot(x, np.sin(x), '-');
plt.plot(x, np.cos(x), '--');
```

```python
#fig.savefig('myfigure.png')
```

## Object-oriented interface:

```python
#First create a grid of plots
# ax will be an array of two axes objects

fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Simple Line Plots:

- **y = f(x)** where x = independent varible, y = dependent variable


```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('classic')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

```python
## Adjusting the plot: Line Colors and Styles

plt.plot(x, np.sin(x), color = 'red');
plt.plot(x, np.sin(x - 1), color = 'g'); #rgbcmyk
plt.plot(x, np.sin(x - 2), color = '0.25') #grayscale
plt.plot(x, np.sin(x - 3), color = '#FFDD44'); #Hex code
plt.plot(x, np.sin(x - 4), color = (1,0.5,0.3)); #rgb tuple
```

```python
## adjust the line style using the linestyle keyword

plt.plot(x, np.sin(x), linestyle = 'solid');
plt.plot(x, np.sin(x + 1), linestyle = 'dashed');
plt.plot(x, np.sin(x + 2), linestyle = 'dashdot');
plt.plot(x, np.sin(x + 3), linestyle = 'dotted');
```

```python
# For short, use the following codes

plt.plot(x, np.sin(x), '--');
plt.plot(x, np.sin(x + 2), '--g');
plt.plot(x, np.sin(x + 3), '-.k');
plt.plot(x, np.sin(x + 4), ':r'); #dotted red
```
