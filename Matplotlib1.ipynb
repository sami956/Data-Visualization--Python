## Introduction to Matplotlib

- Data visualization library buit on NumPy arrays


- Most important feature is its ability to run well with many operating systems and graphics backends


- The best use of Matplotlib differs depending on how you are using it .i.e., In a script, in Ipython terminal
  or in an IPython notebook

## Plotting from an IPython notebook

**%matplotlib** - embed interactive plots within the notebook

**%matplotlob inline** - embed static images of your plot in the notebook (once per kernel/session)
                         (any cell within the notebook that creates a plot will embed a PNG image of the resulting
                         graphic)

```python
%matplotlib inline
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

-  The **figure** (an instance of the class plt.Figure) can be thought of as a single container that containd 
    all the objects representing axes, graphics, text, labels
    
    
- The **axes**(an instance of the class plt.Axes) is a bounding box with ticks and labels, which will eventually
   contain the plot elements that make up our visualization

### Two interfaces to create plots

#### MATLAB-style interface:

```python
x = np.linspace(0,10, 100)
```

```python
plt.figure()  # create a plot figure #stateful object (remembers figure and axis)

# create the first two panels and set current axis
plt.subplot(2,1,1) #(rows, columns, panel number)
plt.plot(x, np.sin(x));

#create the secind panel and set current axis
plt.subplot(2,1,2) 
plt.plot(x, np.cos(x));
```

```python
#Saving Figures to File:

fig = plt.figure()

plt.plot(x, np.sin(x), '-');
plt.plot(x, np.cos(x), '--');
```

```python
#fig.savefig('myfigure.png')
```

## Object-oriented interface:

```python
#First create a grid of plots
# ax will be an array of two axes objects

fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Simple Line Plots:

- **y = f(x)** where x = independent varible, y = dependent variable


```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('classic')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

```python
## Adjusting the plot: Line Colors and Styles

plt.plot(x, np.sin(x), color = 'red');
plt.plot(x, np.sin(x - 1), color = 'g'); #rgbcmyk
plt.plot(x, np.sin(x - 2), color = '0.25') #grayscale
plt.plot(x, np.sin(x - 3), color = '#FFDD44'); #Hex code
plt.plot(x, np.sin(x - 4), color = (1,0.5,0.3)); #rgb tuple
```

```python
## adjust the line style using the linestyle keyword

plt.plot(x, np.sin(x), linestyle = 'solid');
plt.plot(x, np.sin(x + 1), linestyle = 'dashed');
plt.plot(x, np.sin(x + 2), linestyle = 'dashdot');
plt.plot(x, np.sin(x + 3), linestyle = 'dotted');
```

```python
# For short, use the following codes

plt.plot(x, np.sin(x), '--');
plt.plot(x, np.sin(x + 2), '--g');
plt.plot(x, np.sin(x + 3), '-.k');
plt.plot(x, np.sin(x + 4), ':r'); #dotted red
```

### Adjusting the axes limits:

```python

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 1000)
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]); #not plt.axes()
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]); #not plt.axes()
plt.axis('tight');
```

### Labeling Plots:

```python
#Title and axis lables

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel('x')
plt.ylabel('sin(x)');
```

```python
#figure size

plt.figure(figsize=(12,7))

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel('x')
plt.ylabel('sin(x)');
```

```python
# Plot Legend

plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.legend();
```

## Simple Scatter Plots

- Relationship between two or more numerical variables

- Points are used to reperesent data with a dot, circle, or other shape

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('seaborn-whitegrid')
```

```python
#Scatter Plots with plt.plot

x = np.linspace(0,10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black'); #marker parameter
```

```python
for marker in ['o','x', '+', '.', '^', '>', '<', 's', 'd']:
    plt.plot(np.random.rand(5), np.random.rand(5), marker)
```

```python
plt.plot(x, y , 'ok');
```

```python
plt.plot(x, y, '-o', color = 'gray',
                    markersize = 10,
                    markerfacecolor = 'white',
                    markeredgecolor = 'red',
                    markeredgewidth = 2);

plt.ylim(-1.2, 1.2)
```

```python
#Scatter plot with plt.scatter

plt.scatter(x, y, marker = 'o');
```

```python
rng = np.random.RandomState(0) #reproducible results

x = rng.randn(10)
y = rng.randn(10)

colors = rng.rand(10)
sizes = 1000 * rng.rand(10)

plt.scatter(x, y , s = sizes, c = colors, 
            cmap = 'inferno', alpha = 0.7); #alpha = overlaping (transparency)

plt.colorbar(); #show color scale #viridis, inferno, plasma
```

```python
print(x)
print()
print(y)
print()
print(colors)
print()
print(sizes)
```

```python
from sklearn.datasets import load_iris #scikit-learn

iris = load_iris() 

features = iris.data.T

plt.scatter(features[0], features[1], alpha = 0.5,
            cmap = 'viridis', c = iris.target, s = 100 *features[3]);

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

### Histograms

- plot data distribution for a continuous variable(numerical)

```python
plt.style.use('seaborn-white')

data = np.random.randn(500) #10 bins by default

plt.hist(data);
```

```python
plt.hist(data, bins = 30, alpha = 0.5, color = 'green', 
         edgecolor = 'black', histtype = 'stepfilled' );
```

```python
np.random.seed(3)

x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

plt.figure(figsize= (12,7))

kwargs = {'alpha' : 0.3, 'bins' : 40, 'histtype': 'stepfilled'}

plt.hist(x1, label = 'x1', **kwargs)
plt.hist(x2, label = 'x2',**kwargs)
plt.hist(x3,label = 'x3', **kwargs)
plt.legend();
```

```python
counts, bin_edges = np.histogram(data, bins = 40)
print(counts)
```
