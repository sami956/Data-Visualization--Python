### Matplotlib vs Seaborn
- Matplotlibs defaults are not exactly the best choices
- Matplotlibs API is realtively low level
- Matlplotlib was released before pandas, thus is not designed to work with Pandas DataFrames


### Seaborn

- It is an API on top of Matplotlib that offers diverse choices for plot style and color defaults

- Defines high-level functions for common statistical plots, and also integrates functionality provided by
  Pandas DataFrames
  

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
import pandas as pd
```

```python
#Create some data

rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
#Plot the data wiht Matplotlib defaults
plt.plot(x,y);
plt.legend('ABCDEF', loc = 'upper left', ncol = 2);
```

```python
import seaborn as sns
#overwrite Matplotlib's deafult parameters
sns.set()
```

```python
plt.plot(x,y);
plt.legend('ABCDEF', loc = 'upper left', ncol = 2);
```

## Seaborn Plots:

```python
# Histograms

x = np.random.normal(0, 0.8, (1000, 1))
y = np.random.normal(-1, 1, (1000, 1))

data = np.hstack((x,y))

data = pd.DataFrame(data, columns = ['x', 'y'])

kwargs = {'alpha' : 0.5, 'bins' : 40, 'density' : True}

plt.hist(data['x'], label = 'x', **kwargs)
plt.hist(data['y'], label = 'y', **kwargs)
plt.legend();
```

```python
# size(bin) = upperlimit - lowerlimit / no of bins
# Probability Density Function
# Kernel Density Function
```

```python
# Kernel Density Estimation

for col in 'xy':
    sns.kdeplot(data[col], shade = True);
```

```python
# Histograms and KDE combined using distplot
sns.distplot(data['x'])
sns.distplot(data['y']);
```
