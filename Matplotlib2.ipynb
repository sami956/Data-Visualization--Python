## Bar chart

- Visualize data that is organized according to categories as a series of bars, where height of each
  bar represents the values of the data in this category

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('default')
```

```python
areas = pd.read_csv('areas.csv')
areas.head()
```

```python
sorted_areas = areas.sort_values("area (sq. mi)", ascending=False)
```

```python
sorted_areas.set_index('state', inplace = True)
```

```python
sorted_areas.head()
```

```python
top5 = sorted_areas[:5]
top5
```

```python
plt.bar(top5.index, top5['area (sq. mi)']);
```

## Box Plots

- A statistical visualization used to indicate summary statistics of a continuous variable

```python
areas.head()
```

```python
#Box plot

plt.figure(figsize= (12,10))
plt.boxplot([areas['area (sq. mi)']])
plt.xlabel('area (sq. mi)');
```

```python
# 1st quartile(25%)
# 2nd quartile(median 50%)
# 3rd quartile (75%)
# Box(inter-quartile range)
# Gaussian distribution or normal distribution(99% -values lie in the inter-quartile range)
```

## Customize Plot Legends:

```python
%matplotlib inline
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()

ax.plot(x, np.sin(x), '-b', label ='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')

ax.axis('equal')

leg = ax.legend();
```

```python
ax.legend(loc = 'upper left', frameon=False)
fig
```

```python
ax.legend(loc = 'lower center', frameon=False, ncol = 2)
fig
```

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```

## Multile Subplots

- Compare different views of data side by side
- group of smaller axes within a single figure

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('seaborn-white')
```

```python
#plt.axes : Subplots by hand
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])

ax2.plot(x, np.sin(x))
ax2.axis('equal');
```

```python
# Using object-oriented interface

fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                   xticklabels = [], ylim =(-1.2,1.2))

ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],ylim =(-1.2,1.2))

ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```
