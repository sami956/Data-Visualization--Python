```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
# plt.subplots: The whole grid in One Go
# 2x3 grid of subplots
# all axes in the same row share their y-axis scale
# all axes in the same column share their x-axis scale

fig, ax = plt.subplots(2, 3, sharex = 'col', sharey = 'row')
```

```python
ax
```

```python
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i,j)),
                     fontsize = 18, ha = 'center')
        
fig
```

## Text and Annotation

```python
# Add labels to the plot

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal');

style = dict(ha = 'center', size = 12, color = 'b')

ax.text(6.5, 1.2, 'Local Maximum', **style)
ax.text(16, -1.7, 'Local Minimum', **style)


ax.set_title('A Cosine Curve')
ax.set_xlabel('x')
ax.set_ylabel('cos(x)');
```

### Arrows and Annoatation:

```python
plt.style.use('default')

fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)

ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6,1), xytext=(7.5,4),
           arrowprops = dict(facecolor = 'black', shrink = 0.05));

ax.annotate('local minimum', xy=(16,-1), xytext=(2,-6),
           arrowprops = dict(arrowstyle = '->',
                            connectionstyle = 'angle3,angleA=0,angleB=-90'));
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)

ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6,1), xytext=(7.5,4),
            size = 13, ha = 'center', va = 'center',
            bbox=dict(boxstyle = 'round', alpha = 0.3),
            arrowprops = dict(arrowstyle = 'wedge, tail_width=0.5', alpha = 0.3));

ax.annotate('local minimum', xy=(16,-1), xytext=(2,-6),
           arrowprops = dict(arrowstyle = '->',
                            connectionstyle = 'angle3,angleA=0,angleB=-90'));
```

```python
## Major and Minor ticks

ax = plt.axes(xscale='log', yscale='log',
              xlim = (1, pow(10,7)), ylim = (1, pow(10,7)));
```

```python
print(ax.xaxis.get_major_locator())# Major Locator
print(ax.xaxis.get_minor_locator()) 
```

```python
print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
```

```python
# Hiding Ticks and Labels

ax = plt.axes()

ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator());
ax.xaxis.set_major_formatter(plt.NullFormatter());
```

```python
# Reducing or Increasing the number of ticks
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(4,4, sharex = True, sharey = True);
```

```python
for axis in ax.flat:
    axis.xaxis.set_major_locator(plt.MaxNLocator(3))
    axis.yaxis.set_major_locator(plt.MaxNLocator(3))

fig
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randn(1000)
plt.hist(x);
```

```python
#Use a gray background
ax = plt.axes(facecolor = 'gray')
ax.set_axisbelow(True)

#draw solid white grid lines
plt.grid(color = 'w', linestyle = 'solid')

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

ax.tick_params(colors = 'gray')
    
ax.hist(x, color = 'purple');
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1,2, figsize=(12,4))
    
    ax[0].hist(np.random.randn(1000))
    
    for i in range(3):
        ax[1].plot(np.random.rand(10))
        ax[1].legend(['a', 'b', 'c'], loc = 'lower left')
```

```python
hist_and_lines()
```

```python
with plt.style.context('ggplot'):
    hist_and_lines()
```

```python
with plt.style.context('dark_background'):
    hist_and_lines()
```
