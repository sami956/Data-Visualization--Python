## Pair Plots

- Exploring correlations between multidimensional data
- Plot all pairs of values against each other

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.shape
```

```python
# Visualizing the multidimensional
# relationshiop among the samples

sns.pairplot(iris, hue = 'species');
# A pair plot showing the relationships between four variables
```

### Faceted Histograms

- Histograms of subsets

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg.isnull().sum()
```

```python
mpg.dropna(inplace = True)
```

```python
grid = sns.FacetGrid(mpg, row='cylinders', col='origin')
grid.map(plt.hist, 'model_year');
```

```python
iris.head()
```

```python
grid2 = sns.FacetGrid(iris, col = 'species')
grid2.map(plt.scatter, 'petal_length', 'petal_width');
```

## Factor Plots

 - View the distribution of a variable within bins defined by another variable

```python
with sns.axes_style(style = 'ticks'):
    g = sns.catplot('cylinders', 'mpg', 'origin',data=mpg, kind='box', height = 8)
    g.set_axis_labels('Cylinders', 'Miles Per Gallon');
```

### Joint Plots:

- To plot joint distribution between different parameters, along with the associated marginal distributions

```python
with sns.axes_style('white'):
    sns.jointplot('mpg', 'horsepower', data=mpg, kind='hex');
```

```python
# Kernel Density Estimation and regression
with sns.axes_style('white'):
    sns.jointplot('mpg', 'horsepower', data=mpg, kind='reg');
```

### Bar Plots

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
with sns.axes_style('white'):
    sns.catplot('year', data=planets, kind='count', aspect = 2);
```

```python
with sns.axes_style('white'):
    g = sns.catplot('year', data=planets, hue='method', kind='count',
                    aspect = 2, height=8);
    g.set_xticklabels(step=3)
    g.set_ylabels('NUmber of Planets Discovered')
```
